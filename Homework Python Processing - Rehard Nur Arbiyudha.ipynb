```python
import pandas as pd
```

```python
data_analyst = pd.read_csv('data_analyst.csv')
data_analyst
```

#1. Lokasi perusahaan yang paling banyak menyediakan lowongan pekerjaan adalah New York, NY

```python
#Group berdasarkan lokasi dan menghitung jumlah baris
loker = data_analyst.groupby(['Location']).count()
loker.sort_values(by=['Job Title'],ascending= False).head(1)
```

#2. Jumlah lowongan pekerjaan yang memiliki Job Title seputar Reporting adalah 68

```python
#Mencari lowongan pekerjaan yang memiliki Job Title terkait Reporting
job_about_reporting = data_analyst[data_analyst['Job Title'].str.contains("Reporting")].reset_index()
job_about_reporting
```

#3. Rating lowongan pekerjaan pada index 135 adalah 3.3

```python
#Mengeluarkan perusahaan yang Founded bukan dalam bentuk tahun
not_tahun = data_analyst.drop(data_analyst[data_analyst['Founded']==-1].index)
not_tahun
```

```python
#Mengurutkan data frame pada kolom Company Name secara descending dan reset index nya
company_name_desc = not_tahun.sort_values('Company Name',ascending=False).reset_index()
company_name_desc
```

```python
#Mencari Rating lowongan pekerjaan pada baris dengan index 135
company_name_desc.iloc[[135],:]
```

#4. Hasil Rating dari kedua dataframe adalah 4.55

```python
#Filter data frame pada kolom Industry = Investment Banking & Asset Management
investment_job = data_analyst[data_analyst['Industry'].str.contains("Investment")].reset_index()
investment_job
```

```python
#Filter pada dataset awal yang memiliki Rating >=4.5
high_rating_job = data_analyst[data_analyst['Rating']>=4.5].reset_index()
high_rating_job
```

```python
#Mengabungkan dua dataframe tersebut
gabungan_job = pd.concat([investment_job,high_rating_job])
gabungan_job
```

```python
#Mencari nilai rata-rata dari kedua dataframe
gabungan_job['Rating'].mean()
```

#5. Perusahaan nya muncul adalah Tempus Labs\n3.2 berlokasi di Chicago, IL dengan rating 3.2

```python
#Filter dataset mengambil lowongan pekerjaan yang Easy Apply, Founded > 2015, dan Berlokasi di Chicago
easyjob = data_analyst[(data_analyst['Easy Apply'].str.contains("True")
              & ( data_analyst['Founded']>=2015)
              & ( data_analyst['Location'].str.contains("Chicago")))]
#Menampilkan kolom Company Name, Location, dan Rating
easyjob_filter = easyjob.drop(['Unnamed: 0','Job Description','Salary Estimate','Job Title',
              'Headquarters','Size','Founded','Type of ownership','Industry',
              'Sector','Revenue','Competitors','Easy Apply'],axis = 1)
easyjob_filter
```

```python
#Menyusun ulang kolom sesuai urutan Company Name, Location, dan Rating
easyjob_filter[['Company Name','Location','Rating']]
```
